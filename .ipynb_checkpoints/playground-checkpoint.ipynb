```python
import numpy as np

def generate_components(num_components=5, cost_range=(1, 10)):
    """
    Generate a random set of components with their associated costs.
    Args:
        num_components (int): Number of components to generate.
        cost_range (tuple): Min and max range for component costs.

    Returns:
        dict: A dictionary with component IDs as keys and costs as values.
    """
    component_costs = {
        f"Component_{i+1}": np.random.uniform(*cost_range)
        for i in range(num_components)
    }
    return component_costs

def generate_machines(num_machines=3, time_availability_range=(5, 20)):
    """
    Generate a random set of machines with their time availability.
    Args:
        num_machines (int): Number of machines to generate.
        time_availability_range (tuple): Min and max range for time availability in hours.

    Returns:
        dict: A dictionary with machine IDs as keys and time availability as values.
    """
    machine_availability = {
        f"Machine_{i+1}": np.random.uniform(*time_availability_range)
        for i in range(num_machines)
    }
    return machine_availability

def generate_production_times(num_components, num_machines, time_range=(0.5, 5)):
    """
    Generate the time required to produce each component on each machine.
    Args:
        num_components (int): Number of components.
        num_machines (int): Number of machines.
        time_range (tuple): Min and max range for production times in hours.

    Returns:
        np.ndarray: A matrix of shape (num_components, num_machines) with production times.
    """
    production_times = np.random.uniform(*time_range, size=(num_components, num_machines))
    return production_times

def generate_gozinto_factors(num_components, num_products, factor_range=(1, 5)):
    """
    Generate the Gozinto factor matrix linking components to products.
    Args:
        num_components (int): Number of components.
        num_products (int): Number of products.
        factor_range (tuple): Min and max range for Gozinto factors.

    Returns:
        np.ndarray: A matrix of shape (num_components, num_products) with Gozinto factors.
    """
    gozinto_matrix = np.random.randint(*factor_range, size=(num_products, num_components))
    return gozinto_matrix
```

```python
# Example usage
num_components = 5
num_machines = 3
num_products = 2

# Generate components and their costs
components = generate_components(num_components=num_components, cost_range=(1, 10))
print("Components and Costs:")
for k, v in components.items():
    print(f"  {k}: ${v:.2f}")

# Generate machines and their time availability
machines = generate_machines(num_machines=num_machines, time_availability_range=(5, 20))
print("\nMachines and Time Availability (hours):")
for k, v in machines.items():
    print(f"  {k}: {v:.2f} hours")

# Generate production times
production_times = generate_production_times(num_components, num_machines, time_range=(0.5, 5))
print("\nProduction Times (hours per component per machine):")
print(production_times)

# Generate Gozinto factors
gozinto_matrix = generate_gozinto_factors(num_components, num_products, factor_range=(1, 5))
print("\nGozinto Factor Matrix (components required per product):")
print(gozinto_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(27)

# Step 1: Define random parameters
def generate_price_demand_params():
    alpha = np.random.uniform(50, 200)  # Baseline demand
    beta = np.random.uniform(0.5, 5.0)  # Price sensitivity
    return alpha, beta

# Step 2: Generate random prices
n_samples = 1000
prices_A = np.random.uniform(10, 50, n_samples)
prices_B = np.random.uniform(10, 50, n_samples)

# Step 3: Generate demands linked to prices
alpha_A, beta_A = generate_price_demand_params()
alpha_B, beta_B = generate_price_demand_params()

# Introduce random noise
noise_A = np.random.normal(0, 10, n_samples)
noise_B = np.random.normal(0, 10, n_samples)

# Demand functions
demand_A = alpha_A - beta_A * prices_A + noise_A
demand_B = alpha_B - beta_B * prices_B + noise_B

# Ensure demands are non-negative
demand_A = np.maximum(demand_A, 0)
demand_B = np.maximum(demand_B, 0)

# Step 4: Visualize price-demand relationships
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(prices_A, demand_A, alpha=0.5, label="Product A")
plt.title("Price vs Demand (Product A)")
plt.xlabel("Price")
plt.ylabel("Demand")
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(prices_B, demand_B, alpha=0.5, label="Product B", color="orange")
plt.title("Price vs Demand (Product B)")
plt.xlabel("Price")
plt.ylabel("Demand")
plt.legend()

plt.tight_layout()
plt.show()

# Step 5: Print key details
print(f"Product A: alpha={alpha_A:.2f}, beta={beta_A:.2f}")
print(f"Product B: alpha={alpha_B:.2f}, beta={beta_B:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define random parameters
def generate_price_demand_params():
    alpha = np.random.uniform(50, 200)  # Baseline demand
    beta = np.random.uniform(0.5, 5.0)  # Price sensitivity
    return alpha, beta

# Step 2: Generate random prices
n_samples = 1000  # Number of price points
prices_A = np.random.uniform(10, 50, n_samples)
prices_B = np.random.uniform(10, 50, n_samples)

# Step 3: Simulate demand with increasing number of simulations
simulations = [10, 20, 50, 100, 200, 500]
average_demands_A = []
average_demands_B = []

for n_simulations in simulations:
    total_demand_A = np.zeros(n_samples)
    total_demand_B = np.zeros(n_samples)
    
    for _ in range(n_simulations):
        # Generate random parameters
        alpha_A, beta_A = generate_price_demand_params()
        alpha_B, beta_B = generate_price_demand_params()
        
        # Generate random noise
        noise_A = np.random.normal(0, 10, n_samples)
        noise_B = np.random.normal(0, 10, n_samples)
        
        # Calculate demand for this simulation
        demand_A = alpha_A - beta_A * prices_A + noise_A
        demand_B = alpha_B - beta_B * prices_B + noise_B
        
        # Ensure demands are non-negative
        demand_A = np.maximum(demand_A, 0)
        demand_B = np.maximum(demand_B, 0)
        
        # Accumulate demand
        total_demand_A += demand_A
        total_demand_B += demand_B
    
    # Calculate average demand
    avg_demand_A = total_demand_A / n_simulations
    avg_demand_B = total_demand_B / n_simulations
    
    # Store the mean average demand
    average_demands_A.append(np.mean(avg_demand_A))
    average_demands_B.append(np.mean(avg_demand_B))

# Step 4: Plot stability results
plt.figure(figsize=(12, 6))

plt.plot(simulations, average_demands_A, label="Product A", marker="o")
plt.plot(simulations, average_demands_B, label="Product B", marker="o")
plt.title("Stability of Average Demand with Increasing Simulations")
plt.xlabel("Number of Simulations")
plt.ylabel("Mean Average Demand")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(27)

# Step 1: Define parameters
def generate_price_demand_params():
    alpha = np.random.uniform(50, 200)  # Baseline demand
    beta = np.random.uniform(0.5, 5.0)  # Price sensitivity
    return alpha, beta

# Parameters for Products A and B
alpha_A, beta_A = generate_price_demand_params()
alpha_B, beta_B = generate_price_demand_params()

# Cost of components (C_i) for each product
component_costs = [5, 7]  # Component costs for Product A and B
num_components = len(component_costs)

# Selling prices for each product
prices_to_simulate = np.linspace(10, 50, 10)  # 10 prices for simulation

# Number of demand scenarios
n_runs = 200

# Initialize results
expected_revenues = []
component_purchases = []

for i, price_A in enumerate(prices_to_simulate):
    for j, price_B in enumerate(prices_to_simulate):
        # Step 2: Simulate stochastic demand
        noise_A = np.random.normal(0, 10, n_runs)
        noise_B = np.random.normal(0, 10, n_runs)

        demands_A = np.maximum(alpha_A - beta_A * price_A + noise_A, 0)
        demands_B = np.maximum(alpha_B - beta_B * price_B + noise_B, 0)

        # Average demand (E[y_j(w)])
        avg_demand_A = np.mean(demands_A)
        avg_demand_B = np.mean(demands_B)

        # Step 3: Compute revenue and costs
        revenue_A = price_A * avg_demand_A
        revenue_B = price_B * avg_demand_B

        # Component purchases based on demand (sum epsilon(C_i * x_i))
        x_A = avg_demand_A  # Component usage for Product A
        x_B = avg_demand_B  # Component usage for Product B

        total_cost = component_costs[0] * x_A + component_costs[1] * x_B
        total_revenue = revenue_A + revenue_B

        # Net revenue using the lecture formula
        net_revenue = -total_cost + total_revenue

        # Store results
        expected_revenues.append((price_A, price_B, net_revenue))
        component_purchases.append((x_A, x_B))

# Step 4: Visualize Revenue vs Prices
expected_revenues = np.array(expected_revenues)

# Extract prices and net revenues for heatmap
price_A_vals = expected_revenues[:, 0]
price_B_vals = expected_revenues[:, 1]
net_revenue_vals = expected_revenues[:, 2]

# Reshape for 2D heatmap
price_grid_A, price_grid_B = np.meshgrid(prices_to_simulate, prices_to_simulate)
net_revenue_grid = net_revenue_vals.reshape(len(prices_to_simulate), len(prices_to_simulate))

# Plot heatmap
plt.figure(figsize=(8, 6))
plt.contourf(price_grid_A, price_grid_B, net_revenue_grid, cmap="viridis", levels=50)
plt.colorbar(label="Net Revenue")
plt.title("Net Revenue Heatmap")
plt.xlabel("Price of Product A")
plt.ylabel("Price of Product B")
plt.scatter(price_A_vals[np.argmax(net_revenue_vals)], price_B_vals[np.argmax(net_revenue_vals)],
            color='red', label='Optimal Price')
plt.legend()
plt.show()

# Step 5: Print optimal prices and revenues
optimal_idx = np.argmax(net_revenue_vals)
optimal_price_A = price_A_vals[optimal_idx]
optimal_price_B = price_B_vals[optimal_idx]
optimal_net_revenue = net_revenue_vals[optimal_idx]

print(f"Optimal Price for Product A: ${optimal_price_A:.2f}")
print(f"Optimal Price for Product B: ${optimal_price_B:.2f}")
print(f"Maximum Net Revenue: ${optimal_net_revenue:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate problem setup
num_components = 5
num_products = 2

components = generate_components(num_components=num_components, cost_range=(1, 10))
gozinto_matrix = generate_gozinto_factors(num_components, num_products, factor_range=(1, 5))

# Component costs as a numpy array for calculation
component_costs = np.array(list(components.values()))

# Calculate minimum cost to produce each product
product_costs = np.dot(gozinto_matrix.T, component_costs)  # Corrected multiplication

# Define price sets dynamically to ensure they are greater than production costs
price_sets = []
for i in range(10):  # Generate 5 price sets
    price_set = [np.random.uniform(cost + 5, cost + 1000) for cost in product_costs]
    price_sets.append(price_set)

# Display component costs, product costs, and generated price sets
print("Component Costs:")
for i, cost in enumerate(components.values(), 1):
    print(f"  Component_{i}: ${cost:.2f}")

print("\nProduct Costs (Minimum Cost to Produce):")
for i, cost in enumerate(product_costs, 1):
    print(f"  Product_{i}: ${cost:.2f}")

print("\nGenerated Price Sets:")
for idx, price_set in enumerate(price_sets, 1):
    print(f"  Price Set {idx}: {price_set}")

# Number of demand scenarios to evaluate
max_scenarios = 500
scenario_steps = np.arange(10, max_scenarios + 1, 10)

# Initialize results for each price set
revenues_over_scenarios = {tuple(prices): [] for prices in price_sets}

# Evaluate revenue for each price set as the number of scenarios increases
for num_scenarios in scenario_steps:
    # Simulate stochastic demand for products
    demand_distribution = lambda size: np.random.randint(50, 150, size=(size, num_products))
    demand_scenarios = demand_distribution(num_scenarios)
    
    for price_set in price_sets:
        total_revenue = 0  # Initialize total revenue for the given number of scenarios
        
        for demand in demand_scenarios:
            # Step 1: Component usage based on demand and Gozinto matrix
            component_usage = gozinto_matrix @ demand  # Total component needs
            total_costs = np.sum(component_costs * component_usage)
            
            # Step 2: Revenue for this demand scenario
            scenario_revenue = np.sum(np.array(price_set) * demand) - total_costs
            
            # Accumulate revenue
            total_revenue += scenario_revenue
        
        # Store average revenue for this price set and number of scenarios
        revenues_over_scenarios[tuple(price_set)].append(total_revenue / num_scenarios)

# Plot line graph
plt.figure(figsize=(20, 10))
for price_set, revenues in revenues_over_scenarios.items():
    plt.plot(scenario_steps, revenues, label=f"Prices: {np.round(price_set, 2)}")

plt.title("Net Revenue vs Number of Scenarios")
plt.xlabel("Number of Scenarios")
plt.ylabel("Net Revenue")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

def generate_stochastic_demand(price_sets, num_scenarios, seasonality_period=365, noise_std=10):
    """
    Generate stochastic demand scenarios linking prices and demands with enhanced factors.
    Args:
        price_sets (list): List of price sets.
        num_scenarios (int): Number of scenarios to generate.
        seasonality_period (int): Period for seasonal demand fluctuations (e.g., 365 for yearly).
        noise_std (float): Standard deviation of random noise.

    Returns:
        numpy.ndarray: Demand scenarios for all products and scenarios.
    """
    num_price_sets = len(price_sets)
    num_products = len(price_sets[0])
    demands = np.zeros((num_price_sets, num_scenarios, num_products))
    
    # Generate model parameters
    alphas = np.random.uniform(50, 200, size=num_products)  # Baseline demand
    betas = np.random.uniform(0.5, 5.0, size=num_products)  # Price sensitivity
    gammas = np.random.uniform(-1, 1, size=(num_products, num_products))  # Cross-price elasticities
    np.fill_diagonal(gammas, 0)  # Self-elasticity does not apply for cross-prices
    deltas = np.random.uniform(0, 50, size=num_products)  # Seasonal amplitudes

    for price_set_idx, price_set in enumerate(price_sets):
        for scenario in range(num_scenarios):
            # Simulate time factor for seasonality
            t = np.random.randint(0, seasonality_period)
            seasonality = deltas * np.sin(2 * np.pi * t / seasonality_period)

            for product_idx in range(num_products):
                # Price impact
                own_price_effect = -betas[product_idx] * price_set[product_idx]

                # Cross-price effects
                cross_price_effect = np.sum([
                    gammas[product_idx, other_idx] * price_set[other_idx]
                    for other_idx in range(num_products) if other_idx != product_idx
                ])

                # Random noise
                noise = np.random.normal(0, noise_std)

                # Calculate demand
                demands[price_set_idx, scenario, product_idx] = max(
                    0, alphas[product_idx] + own_price_effect + cross_price_effect + seasonality[product_idx] + noise
                )
    
    return demands
```

```python
# Example price sets
price_sets = [
    [50, 60],  # Price set 1
    [55, 65],  # Price set 2
    [45, 55]   # Price set 3
]

# Generate demands
num_scenarios = 100
demands = generate_stochastic_demand(price_sets, num_scenarios)

# View results for one price set
print("Demand Scenarios for Price Set 1:")
print(demands[0])
```
